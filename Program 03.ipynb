```python
import numpy as np 
import matplotlib.pyplot as plt 
import math
```

```python
x = np.matrix('0 0 1 1;0 1 0 1')

t = np.matrix('0 1 1 0')
cost = []
count = 0
it = []

w1 = np.matrix('0.2 0.3;0.1 -0.2;-0.5 0.4')
print(w1)

w2 = np.matrix('0.1 -0.2 0.4;-0.5 0.2 -0.1')
print(w2)

w3 = np.matrix('0.1;-0.2')
print(w3)

b1 = np.matrix('0;0;0')
print(b1)

b2 = np.matrix('0;0')
print(b2)

b3 = np.matrix('0')
print(b3)

for i in range(0,4000):
    z1 = w1 * x + b1
    print(z1)

def sigmoid(z):
    p = 1/(1+(np.exp(-z)))
    return p

def differ(z):
    G = sigmoid(z)
    res  = np.multiply(G,(1-G))
    return res

a1 = sigmoid(z1)
print(a1)
z2 = np.multiply(w2,a1) + b2
print(z2)

a2 = sigmoid(z2)
print(a2)

z3 = np.multiply(w3,a2) + b3
print(z3)

a3 = sigmoid(z3)

J = -np.sum((np.multiply(t,np.log(a2))) + (np.multiply((1-t),np.log(1-a2))))
print(J)

diffZ3 = a3 - t
print(diffZ3)

a2t = a2.transpose()
diffW3 = diffZ3 * a2t
print(diffW3)

diffB3 = diffZ3
W3t = w3.transpose()

diffA2 = W3t * diffZ3
print(diffA2)

diffZ2 = np.multiply(diffA2,differ(a2))
print(diffZ2)

alt = a1.transpose()
diffW2 = diffZ2 * alt
print(diffW2)

diffB2 = diffW2
W2t = w2.transpose()
diffA1 = W2t * diffZ2
print(diffA1)

diffZ1 = np.multiply(diffA1,differ(a1))
print(diffZ1)

diffB1 = diffZ1
a0t = x.transpose()
diffW1 = diffZ1 * a0t
print(diffW1)
print(diffB1)

w1 = w1 - np.multiply(0.1,diffW1)
b1 = b1 - np.multiply(0.1,diffB1)
w2 = w2 - np.multiply(0.1,diffW2)
b2 = b2 - np.multiply(0.1,diffB2)
w3 = w3 - np.multiply(0.1,diffW3)
b3 = b3 - np.multiply(0.1,diffB3)

cost.append(J)
count = count+1
it.append(count)

plt.plot(it,cost)
plt.xlabel('iterations')
plt.ylabel('cost')
plt.show()
```
